```python
# 将薪资单位统一为 万/年
import math
def getSalary1(s):
    if '以'  in s:
        if '千' in s:
            return float(s.split('千')[0])*1.2
        if '万' in s:
            if '年' in s :
                return float(s.split('万')[0])
            if '月' in s:
                return float(s.split('万')[0])*12
    if '万/年' in s:
        return float(s.split('-')[0])
    if '万/月' in s :
        return float(s.split('-')[0])*12
    if '千/月' in s:
        return float(s.split('-')[0])*1.2
    if '元/天' in s: 
        s = s.split('元')[0]
        return float(s.split('-')[0])*0.024
    return float(s.split('-')[0])
```

```python
ai_salaryValue = excel_AI['薪资'].dropna(axis=0).map(getSalary1)
ai_salaryValue
```

```python
# 得到AI的最低平均工资
sum1 = 0
for value in ai_salaryValue:
    sum1+=value
print(sum1/87)
avg1 = sum1/87
```
